```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv("C:\\Users\\siddh\\OneDrive\\Desktop\\DSBDA\\iriss.csv")
```

```python
df_x=df.iloc[:,0:4].values
```

```python
df_y=df.iloc[:,4].values
```

```python
df_y
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
df_y=le.fit_transform(df_y)
```

```python
df_y
```

```python
x_train,x_test,y_train,y_test=train_test_split(df_x,df_y,test_size=0.3,random_state=0)
```

```python
x_train
```

```python
y_train
```

```python
x_test
```

```python
y_test
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
reg=GaussianNB()
```

```python
reg.fit(x_train,y_train)
```

```python
y_predict=reg.predict(x_test)
```

```python
y_predict
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm=confusion_matrix(y_test,y_predict)
```

```python
cm
```

```python
from sklearn.metrics import accuracy_score
```

```python
print(accuracy_score(y_test,y_predict)*100)
```

```python
from sklearn.metrics import precision_score
```

```python
print(precision_score(y_test,y_predict,average="micro")*100)
```

```python
from sklearn.metrics import recall_score
```

```python
print(recall_score(y_test,y_predict,average="micro")*100)
```
