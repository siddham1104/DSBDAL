```python
import nltk
```

```python
from nltk.tokenize import word_tokenize,sent_tokenize
```

```python
nltk.download('punkt')
```

```python
document='''
Natural language processing (NLP) is a subfield of artificial intelligence 
and linguistics concerned with the interactions between computers and human 
(language) text. NLP, a component of AI, focuses on the ability of computers 
to understand, interpret, and generate human language in a valuable way. 
NLP techniques include tokenization, POS tagging, stop words removal, 
stemming, and lemmatization.
'''
```
